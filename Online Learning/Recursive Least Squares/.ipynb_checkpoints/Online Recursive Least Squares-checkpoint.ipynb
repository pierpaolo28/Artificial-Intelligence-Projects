# Online Recursive Least Squares

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn import preprocessing
from sklearn import linear_model
from sklearn.tree import DecisionTreeRegressor
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.neural_network import MLPRegressor
```

```python
%matplotlib inline
%matplotlib notebook
import random
from sklearn.model_selection import learning_curve
from sklearn.linear_model import Ridge
import matplotlib.pyplot as plt
import seaborn as sns
from ipywidgets import *
import warnings
warnings.filterwarnings('ignore')
sns.set()
```

## Helper Functions

```python
def values_scatter(name, yTarget, yEst):
    plt.figure(num=None, figsize=(6, 6), dpi=200, facecolor='w', edgecolor='k')
    plt.scatter(yTarget, yEst)
    plt.title(name + " (Values Prediction)", fontsize=20)
    plt.xlabel("True Values", fontsize=18)
    plt.ylabel("Predicted Values", fontsize=18)
    plt.tight_layout()
    plt.savefig(name + " Values Prediction.png", format='png', dpi=1200)
    
def compare_series(yTarget, yEst):
    plt.figure(num=None, figsize=(8, 4), dpi=200, facecolor='w', edgecolor='k')
    plt.title("Total Orders Records", fontsize=20)
    plt.xlabel("Time", fontsize=18)
    plt.ylabel("Number of Orders", fontsize=18)
    plt.plot(yEst, label='Estimated Values')
    plt.plot(yTarget, label='Actual Values')
    plt.legend(fontsize=18)
    plt.tight_layout()
    plt.savefig("Preds vs Actual.png", format='png', dpi=1200)
```

## Data Analysis

```python
df = pd.read_csv('Orders.csv', ';')
X = df.drop(['Target (Total orders)'], axis=1).values
X = StandardScaler().fit_transform(X)
yTarget = df['Target (Total orders)']
# Creating a scaler variable in order to scale back our future
# model predictions
scaler = StandardScaler()
yTarget = scaler.fit_transform(yTarget.values.reshape(-1, 1))
df.head()
```

```python
def ml_comparison(X, yTarget):
    ml_models = [linear_model.SGDRegressor(), DecisionTreeRegressor(random_state=0), 
                 GaussianProcessRegressor(random_state=0),
                KNeighborsRegressor(n_neighbors=2), linear_model.BayesianRidge(), 
                 MLPRegressor(hidden_layer_sizes=(30,30,30))]

    for i in ml_models:
        regressor = i 
        regressor.fit(X, yTarget)
        yEst2 = regressor.predict(X)
        print(str(i).split('(')[0] + " Residual Error: ", round(np.linalg.norm(yEst2 - yTarget), 3))

ml_comparison(X, yTarget)
```

```python
def rlr(X, lmd):
    E_plot = np.zeros((len(X), 1))
    w = np.zeros((X.shape[1], 1))
    Rn_inv = np.eye(X.shape[1])
    for i in range(len(X)):
        x = np.reshape(X[i], (X[i].shape[0], 1))
        num = ((1/lmd)*Rn_inv@x)
        dem = (1 + (1/lmd)*x.T@Rn_inv@x)
        kn = num/dem
        error = yTarget[i] - w.T@x
        w = w + kn @ error
        Rn_inv = (1/lmd)*Rn_inv - (1/lmd)*kn*x.T*Rn_inv
        E_plot[i] = np.linalg.norm(X @ w - yTarget)
    return E_plot, w

E, w_rlr = rlr(X, lmd=0.99)
yEst = X @ w_rlr
print("Residual Error: ", round(np.linalg.norm(yEst - yTarget), 3))
values_scatter('Recursive Least Squares', yTarget, yEst)
```

```python
y_plot_target = scaler.inverse_transform(yTarget)
y_plot_est = scaler.inverse_transform(yEst)
compare_series(y_plot_target, y_plot_est)
```

```python
def mbgd(X, lRate, b_size):
    Eplot = np.zeros((len(X), 1))
    batch_size = len(yTarget)//b_size
    wIter = w0
    for itern in range(len(X)):
        rand_sample = random.randint(0, len(X)-1)
        x_rand_sample = X[rand_sample:batch_size+rand_sample, :]
        y_rand_sample = yTarget[rand_sample:batch_size+rand_sample,:]
        yPred = x_rand_sample @ wIter
        wIter = wIter - lRate  * x_rand_sample.T @ (yPred - y_rand_sample)
        Eplot[itern] = np.linalg.norm(X @ wIter - yTarget)
    
    return wIter, Eplot

# Initial guess and error
w0 = np.random.randn(X[0].size,1)
wEst, Embgd = mbgd(X, 0.0095, 16)
yEst = X @ wEst
# Error From the Model
print("Residual Error: ", round(np.linalg.norm(yEst - yTarget), 3))
values_scatter('MBGD Linear Regression', yTarget, yEst)
```

```python
plt.figure(num=None, figsize=(8, 6), dpi=200, facecolor='w', edgecolor='k')
plt.plot(E, label='RLS')
plt.plot(Embgd, label='Mini Batch GD')
plt.xlabel("Iteration Number", fontsize=18)
plt.ylabel("Residual Error", fontsize=18)
plt.legend(fontsize=18)
plt.title("Linear Regression Methods Comparison", fontsize=20)
plt.tight_layout()
plt.savefig("Linear Regression Methods Comparison.png", format='png', dpi=1200)
plt.show();
```

```python
def update(lmd):
    Eplot = np.zeros((len(X), 1))
    w = np.zeros((X.shape[1], 1))
    Rn_inv = np.eye(X.shape[1])
    for i in range(len(X)):
        x = np.reshape(X[i], (X[i].shape[0], 1))
        num = ((1/lmd)*Rn_inv@x)
        dem = (1 + (1/lmd)*x.T@Rn_inv@x)
        kn = num/dem
        error = yTarget[i] - w.T@x
        w = w + kn @ error
        Rn_inv = (1/lmd)*Rn_inv - (1/lmd)*kn*x.T*Rn_inv
        Eplot[i] = np.linalg.norm(X @ w - yTarget)

    plt.figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
    plt.plot(Eplot)
    plt.xlabel("Iteration Number", fontsize=18)
    plt.ylabel("Residual Error", fontsize=18)
    plt.ylim((0,10))
    plt.title("Linear Regression Methods Comparison", fontsize=20)
    plt.show()

interactive_plot = interactive(update, lmd = (0.0001,10.9999));
interactive_plot.layout.height = '600px'
interactive_plot 
```
