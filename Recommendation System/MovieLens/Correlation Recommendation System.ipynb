# Correlation Recommendation System - MovieLens

For this recommendation system has been used the correlation between the ratings of the movie. A matrix has been  created where each column is a movie name and each row contains instead the rating assigned an user to that movie.

## Preprocessing

```python
import numpy as np  
import pandas as pd
import matplotlib.pyplot as plt  
import seaborn as sns
```

```python
df = pd.read_csv("ratings.csv")  
df.head()
```

```python
df2 = pd.read_csv("movies.csv")  
df2.head()  
```

```python
df3 = pd.merge(df, df2, on='movieId') 
df3.head()
```

```python
df3.groupby('title')['rating'].mean().head() 
```

```python
df3.groupby('title')['rating'].mean().sort_values(ascending=False).head()  
```

```python
df3.groupby('title')['rating'].count().sort_values(ascending=False).head()  
```

```python
mean_df = pd.DataFrame(df3.groupby('title')['rating'].mean())  
mean_df['rating_counts'] = pd.DataFrame(df3.groupby('title')['rating'].count())
mean_df.head() 
```

## Correlation

```python
df4 = df3.pivot_table(index='userId', columns='title', values='rating') 
df4.head() 
```

## Evaluation

```python
ratings = df4['Forrest Gump (1994)'] 
movies_like = df4.corrwith(ratings)
corr_values = pd.DataFrame(movies_like, columns=['Correlation'])  
corr_values.dropna(inplace=True)  
corr_values.sort_values('Correlation', ascending=False)
corr_values = corr_values.join(mean_df['rating_counts']) 
corr_values[corr_values['rating_counts']>50].sort_values('Correlation', ascending=False).head() 
```
